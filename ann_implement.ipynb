```python
import pandas as pd
data = pd.read_csv("Churn_Modelling.csv")
data
```

```python
data = data.drop(['RowNumber','CustomerId','Surname'],axis=1)
data.head()
```

```python
from sklearn.preprocessing import LabelEncoder
gender_encoder = LabelEncoder()
data["Gender"] = gender_encoder.fit_transform(data["Gender"])
```

```python
data
```

```python
from sklearn.preprocessing import OneHotEncoder
onehot_encode_geo = OneHotEncoder()
encoder_geo = onehot_encode_geo.fit_transform(data[["Geography"]])
```

```python
onehot_encode_geo.get_feature_names_out(["Geography"])
```

```python
geo_encoded_df = pd.DataFrame(encoder_geo.toarray(),columns=onehot_encode_geo.get_feature_names_out(["Geography"]))
```

```python
data = pd.concat([data.drop("Geography",axis=1),geo_encoded_df],axis=1)
```

```python
data
```

```python
import pickle
with open("gender_encode.pkl",'wb') as file:
    pickle.dump(gender_encoder,file)
with open("onehot_encode_geo.pkl",'wb') as file:
    pickle.dump(onehot_encode_geo,file)
```

```python
x = data.drop("Exited",axis=1)
y = data["Exited"]
```

```python
from sklearn.model_selection import train_test_split
x_train,x_test,y_train,y_test = train_test_split(x,y,train_size=0.1,random_state=42)
```

```python
from sklearn.preprocessing import StandardScaler
scaler = StandardScaler()
x_train = scaler.fit_transform(x_train)
x_test = scaler.transform(x_test)
```

```python
import pickle
with open("scaler.pkl","wb") as file:
    pickle.dump(scaler,file)
```

```python
import tensorflow
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Dense
```

```python
model = Sequential([
    Dense(64,activation='relu',input_shape=(x_train.shape[1],)),
    Dense(32,activation='relu'),
    Dense(1,activation='sigmoid')
])
```

```python
model.summary()
```

```python
import tensorflow
opt = tensorflow.keras.optimizers.Adam(learning_rate=0.1)

model.compile(optimizer=opt,loss="binary_crossentropy",metrics=['accuracy'])
```

```python
from tensorflow.keras.callbacks import EarlyStopping, TensorBoard
```

```python
import datetime
log_dir = "logs/fit/" + datetime.datetime.now().strftime("%Y%m%d-%H%M%S")
tensorboard_callback = TensorBoard(log_dir=log_dir,histogram_freq=1)
```

```python
earlystopping_callback = EarlyStopping(monitor="val_loss",patience=10,restore_best_weights=True)
```

```python
history = model.fit(
    x_train,y_train,
    validation_data = (x_test,y_test),epochs=100,
    callbacks=(tensorboard_callback,earlystopping_callback)

)
```

```python
model.save("model.h5")
```
